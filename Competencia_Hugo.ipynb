```python
import pandas as pd 
from sklearn.ensemble import RandomForestClassifier
```

## **Carga de Datos**

```python
train = pd.read_csv('https://raw.githubusercontent.com/zack0712/Clase-Machine-Learning/main/datos_competencia/train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/zack0712/Clase-Machine-Learning/main/datos_competencia/test.csv')
```

```python
train.head(2)
```

```python
test.head(2)
```

## **Preparación Set de Entrenamiento**

```python
train.info()
```

```python
for columna in test.columns:
  print(columna)
  print(test[columna].unique())
  print(' ')
```

### **Corrección Missing Values Train set**

```python
for columna in train.columns:
  train[columna] = train[columna].fillna(train[columna].mean())
```

```python
train.head()
```

## **Correción Missing Values Test Set**

```python
for columna in test.columns:
  test[columna] = test[columna].fillna(test[columna].mean())
```

## **Modelo**

```python
y = train['target']
X = train.drop(columns=['target','id'])
X_test = test.drop(columns=['id'])
id = test['id']

clf = RandomForestClassifier(max_depth=2, random_state=777)
clf.fit(X,y)

y_pred = clf.predict(X_test)
```

```python
resultados = pd.DataFrame({'id':id,'Predicted':y_pred})
```
